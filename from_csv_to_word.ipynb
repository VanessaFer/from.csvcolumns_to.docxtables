```python
#Import Python libraries
import pandas as pd
import numpy as np
import docx
from docx import Document
from docx.shared import Cm , Pt
```

```python
#Read the csv file with Pandas
df = pd.read_csv('/content/gdrive/MyDrive/Colab_Notebooks/Stele/Stele con tutto.csv')
df
```

```python
#Rename column names in order to normalise it
df =(df.rename(columns={'N_ d_inventario ordinato': 'Inv', 'Titolo, Descrizione': 'Titolo', 'Materiale, tecnica': 'Materiale', 'prov_': 'Prov', 
                            'Acquisito da:': 'Acquisizione', 'Anno pubbl_': 'Anno', 'Casa editrice': 'Casa_editrice'}))

print(df)
```

```python
#Drop the duplicate data
new_df = df.drop_duplicates(subset=['Inv'])
new_df
```

```python
#Drop the columns I don't need
df2 = new_df.drop(columns=["Autore", "Libro", "Anno", "Casa_editrice", "Pg"], axis = 1)

df2
```

```python
#Replace the null values
df3 = df2.replace(np.nan, '', regex=True)
```

```python
#Creation of the docx document
document = docx.Document()

font = document.styles['Normal'].font
font.name = 'Times New Roman'
font.size = Pt(12)

sections = document.sections
for section in sections:
    section.top_margin = Cm(1.25)
    section.bottom_margin = Cm(1.25)
    section.left_margin = Cm(1.75)
    section.right_margin = Cm(1.75)


for idx,col in df3.iterrows():
  t = pd.DataFrame(col)
  table = document.add_table(df3.shape[-1], 2)
  table.style = 'TableGrid'
  for row in table.rows:
      row.height = Cm(1.0)
  for j in range(df3.shape[-1]): #originale -1
      table.cell(j, 0).text = df3.columns[j]
      #for h in range(t.shape[0]):
      table.cell(j, 1).text = str(t.iloc[j,0])
  document.add_page_break() 
```

```python
#Save the Word document
document.save('MEStelae.docx')
from google.colab import files
files.download('MEStelae.docx')
```
